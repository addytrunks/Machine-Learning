# Simple Linear Regression

Regression = Prediction based on the dataset

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
print(X_train)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

# Basically training the modal to predict expected salaries
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
#If the scattered plots are close to the blue line, it means that the predicted values WILL be precise and accurate
plt.scatter(X_train,y_train,color="red")

#The blue line corresponds to the regression line, from which you can predict salaries
plt.plot(X_train,regressor.predict(X_train))
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

## Visualizing the Test Set

```python
plt.scatter(X_test,y_test,color="red")
plt.plot(X_test,regressor.predict(X_test))
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```
