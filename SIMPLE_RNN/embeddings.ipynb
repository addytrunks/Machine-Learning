```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
voc_size = 10000
```

```python
one_hot_rep = [one_hot(words,voc_size) for words in sent]
one_hot_rep
```

## Word Embedding Representation

```python
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
'''
The size of the inputs must be same when training a DL model, so when you have inputs of different shapes,you have
to add padding(0s) either to the front (post) or to the back (pre) to make the shapes equal.
'''
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
```

```python
embedded_docs
```

```python
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
